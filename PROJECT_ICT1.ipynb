```python
import os
from datetime import datetime

BASE_DIR = os.getcwd()
FILE_NAME = os.path.join(BASE_DIR, "expenses.txt")


def add_expense():
    category = input("Enter category: ").strip()

    try:
        amount = float(input("Enter amount: "))
    except ValueError:
        print("Invalid amount. Please enter a number.")
        return

    date = datetime.now().strftime("%Y-%m-%d")

    with open(FILE_NAME, "a") as file:
        file.write(f"{date},{category},{amount}\n")

    print("Expense added successfully.")


def view_expenses():
    try:
        with open(FILE_NAME, "r") as file:
            records = file.readlines()
    except FileNotFoundError:
        print("No expense file found.")
        return

    if not records:
        print("No expenses to display.")
        return

    print("\nAll Expenses")
    print("-" * 40)

    for index, record in enumerate(records, start=1):
        parts = record.strip().split(",")

        if len(parts) == 3:
            date, category, amount = parts
            try:
                amount = float(amount)
                print(f"{index}. {date} | {category} | ₹{amount:.2f}")
            except ValueError:
                continue


def calculate_total():
    total = 0

    try:
        with open(FILE_NAME, "r") as file:
            for record in file:
                parts = record.strip().split(",")
                if len(parts) == 3:
                    try:
                        total += float(parts[2])
                    except ValueError:
                        continue
    except FileNotFoundError:
        print("No expense file found.")
        return

    print(f"\nTotal expense: ₹{total:.2f}")


def analyze_spending():
    category_total = {}

    try:
        with open(FILE_NAME, "r") as file:
            records = file.readlines()
    except FileNotFoundError:
        print("No expense file found.")
        return

    if not records:
        print("No expenses to analyze.")
        return

    for record in records:
        parts = record.strip().split(",")
        if len(parts) == 3:
            date, category, amount = parts
            try:
                amount = float(amount)
                category_total[category] = category_total.get(category, 0) + amount
            except ValueError:
                continue

    print("\nCategory-wise Spending")
    print("-" * 40)

    for category, total in category_total.items():
        print(f"{category}: ₹{total:.2f}")


def main():
    while True:
        print("\n" + "=" * 40)
        print("       EXPENSE TRACKER MENU")
        print("=" * 40)
        print("1. Add Expense")
        print("2. View Expenses")
        print("3. Total Expense")
        print("4. Analyze Spending")
        print("5. Exit")

        choice = input("Enter your choice: ").strip()

        if choice == "1":
            add_expense()
        elif choice == "2":
            view_expenses()
        elif choice == "3":
            calculate_total()
        elif choice == "4":
            analyze_spending()
        elif choice == "5":
            print("Exiting program. Goodbye 👋")
            break
        else:
            print("Invalid choice. Please try again.")


if __name__ == "__main__":
    main()
```
